# ** `scipy.spatial.distance` 주요 메서드 정리**
모든 메서드는 벡터(또는 행렬) 간 거리를 계산하는 데 사용됩니다.

## **1. 유클리드 기반 거리 (Euclidean-based Distances)**
| 메서드 | 설명 | 공식 |
|--------|------|------|
| `euclidean(u, v)` | 유클리드 거리 (직선 거리) | \( d = \sqrt{∑u_i - v_i)^2} \) |
| `sqeuclidean(u, v)` | 제곱 유클리드 거리 (제곱 값만 반환) | \( d = ∑ (u_i - v_i)^2 \) |

---

## **2. 맨해튼 및 민코프스키 거리 (Minkowski-based Distances)**
| 메서드 | 설명 | 공식 |
|--------|------|------|
| `cityblock(u, v)` | 맨해튼 거리 (택시 거리) | \( d = \sum |u_i - v_i| \) |
| `minkowski(u, v, p)` | 민코프스키 거리 (p 값을 조절) | \( d = \left( \sum |u_i - v_i|^p \right)^{\frac{1}{p}} \) |
| `chebyshev(u, v)` | 쳬비쇼프 거리 (최대 차이값) | \( d = \max |u_i - v_i| \) |

---

## **3. 마할라노비스 거리 (Mahalanobis Distance)**
| 메서드 | 설명 | 공식 |
|--------|------|------|
| `mahalanobis(u, v, VI)` | 마할라노비스 거리 (공분산 행렬 고려) | \( d = \sqrt{(u - v)^T S^{-1} (u - v)} \) |

**파라미터**
- `VI`: 공분산 행렬의 역행렬 (\( S^{-1} \))

---

## **4. 확률 기반 거리 (Probability-based Distances)**
| 메서드 | 설명 | 공식 |
|--------|------|------|
| `cosine(u, v)` | 코사인 거리 (유사도와 반대 개념) | \( d = 1 - \frac{u \cdot v}{\|u\|\|v\|} \) |
| `correlation(u, v)` | 피어슨 상관 거리 | \( d = 1 - \frac{\sum (u_i - \bar{u})(v_i - \bar{v})}{\sqrt{\sum (u_i - \bar{u})^2} \sqrt{\sum (v_i - \bar{v})^2}} \) |
| `jensenshannon(p, q)` | 젠슨-섀넌 거리 (확률 분포 차이) | \( d = \sqrt{\frac{1}{2} D_{KL}(p||m) + \frac{1}{2} D_{KL}(q||m)} \) |
| `kulsinski(u, v)` | 쿨신스키 거리 (이진 데이터) | \( d = \frac{c_{10}}{c_{10} + c_{01} + c_{11}} \) |

---

## **5. 벡터 기반 거리 (Binary Distances, Set-based)**
| 메서드 | 설명 | 공식 |
|--------|------|------|
| `hamming(u, v)` | 해밍 거리 (이진 데이터에서 다른 값 개수) | \( d = \frac{1}{n} \sum \mathbb{1}(u_i \neq v_i) \) |
| `jaccard(u, v)` | 자카드 거리 (집합 차이) | \( d = 1 - \frac{|A \cap B|}{|A \cup B|} \) |
| `dice(u, v)` | 다이스 거리 (자카드 거리의 변형) | \( d = 1 - \frac{2 |A \cap B|}{|A| + |B|} \) |
| `rogerstanimoto(u, v)` | 로저스-타니모토 거리 (이진 벡터 비교) | \( d = \frac{2(c_{01} + c_{10})}{c_{00} + c_{11} + 2(c_{01} + c_{10})} \) |
| `russellrao(u, v)` | 러셀-라오 거리 (두 벡터의 1의 개수 차이) | \( d = 1 - \frac{c_{11}}{n} \) |
| `sokalmichener(u, v)` | 소칼-미크너 거리 (이진 데이터 거리 척도) | \( d = \frac{2(c_{01} + c_{10})}{2(c_{01} + c_{10}) + c_{00} + c_{11}} \) |
| `sokalsneath(u, v)` | 소칼-스니스 거리 (군집 분석에서 사용) | \( d = \frac{c_{01} + c_{10}}{c_{01} + c_{10} + 2c_{11}} \) |
| `yule(u, v)` | 율 거리 (이진 데이터 간 연관성) | \( d = \frac{c_{01} c_{10} - c_{00} c_{11}}{c_{01} c_{10} + c_{00} c_{11}} \) |

---

## **6. 거리 행렬 계산 (Pairwise Distance Matrix Calculation)**
| 메서드 | 설명 |
|--------|------|
| `pdist(X, metric='euclidean')` | 행렬 X에서 모든 데이터 쌍에 대한 거리 계산 |
| `cdist(XA, XB, metric='euclidean')` | 두 개의 행렬 \( XA \)와 \( XB \) 간의 거리 계산 |
| `squareform(Y)` | 거리 벡터를 거리 행렬로 변환 |

```python
import numpy as np
## scipy.spatial.distance 모듈에서 distance 함수 불러오기
from scipy.spatial import distance
```

```python
## 벡터 정의
p = np.array([1, 2])
q = np.array([5, 10])
```

```python
p
```

```python
q
```

```python
## 유클리드 거리 계산 (L2 norm)
np.linalg.norm(p - q)
```

```python
## 맨해튼 거리 (L1 norm)
np.sum(np.abs(p - q))
```

```python
## 쳬비셰프 거리
np.max(np.abs(p - q))
```

```python
## 민코프스키 거리
def min_dis(p, q, p_value):
  return np.sum(np.abs(p-q) ** p_value) * (1/p_value)

## L1 norm = Manhattan distance
min_dis(p, q, 1)
```

```python
## L2 norm = Euclidean distance
min_dis(p, q, 2)
```

```python
## 코사인 유사도
## 두 벡터의 내적 / 벡터 p의 크기*벡터 q의 크기
np.dot(p, q) / (np.linalg.norm(p) * np.linalg.norm(q))
```

## 지도에 좌표를 찍어서 확인해보자

```python
## Euclidean distance 계산 함수 정의
def cal_euc_dis(row1, row2):
  pnt1 = np.array([row1['latitude'], row1['longitude']])
  pnt2 = np.array([row2['latitude'], row2['longitude']])
  return np.linalg.norm(pnt1 - pnt2)
```

```python
## 지도에 기준점이 필요함
## 기준점은 평균으로 잡고 각 거리를 계산

## 지도 중심점 계산
map_center = [df_gn['latitude'].mean(), df['longitude'].mean()]

## 지도 위치 시각화
## 지도 객체 생성 (folium)
m = folium.Map(location=map_center, zoom_start=10)
```

```python
## 모든 위치에 있는 스타벅스 강남구 매장들 마커 찍어서 거리 계산하면 된다.

## 모든 매장에 마커 찍기
for idx, row in df_gn.iterrows():
  folium.Marker(location = [row['latitude'], row['longitude']], popup=row['name']).add_to(m)

## 첫 번째 출발점 지정
start_point = df_gn.iloc[0]

## 첫 번째 매장과 나머지 매장의 거리 계산 & 선 연결
for idx, row in df_gn.iterrows():
  if start_point['name'] != row['name']:
    dis = cal_euc_dis(start_point, row)
    folium.PolyLine(locations = [[start_point['latitude'], start_point['longitude']],
                                 [row['latitude'], row['longitude']]]).add_to(m)
```
