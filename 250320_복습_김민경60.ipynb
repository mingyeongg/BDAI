### 불린값 추출

- df[df[]] 불린 문법
- df[True] = 참값이 출력이된다
- True 조건을 만들고 데이터프레임으로 추출하면 -> 해당 기준의 값들이 추출된다.
- df[df['조건1']], 조건 1이 참인 값들이 추출이 된다.
- 다중조건인 경우?
- df[df[(조건1)&(조건2)...&(조건n)]]

```python
# 불린값 추출

tt['age'] > 30
```

```python
# df[df[...]] ...을 만족하는 값이 True인 정보들 출력

tt[tt['age'] > 30]
```

```python
## 2개의 조건
## 2개 같은 기준으로 출력된 문법

print(tt.query("age > 30 & sex=='female' & pclass == 1")['fare'].mean())
print(tt[(tt['age']>30) & (tt['sex']=='female')& (tt['pclass']==1)]['fare'].mean())
```

- groupby - 요약 집계 -> agg함수 같이 데이터를 집계해서 사용한다.
    - 데이터를 요약할 때 많이 사용하실 것
    - sex에 따른 age, fare
    - male - age , fare 집계할 수 있다. -> 요약통계치가 필요하다. 통계함수랑 같이 많이 사용한다.
    - female - age, fare 집계할 수 있다. -> 요약통계치가 필요하다. 통계함수랑 같이 많이 사용한다.
    
- 데이터프레임.groupby(그룹을묶을 컬럼)
- 내가 원하는 그룹의 데이터프레임만 추출
    - 데이터프레임.get_group(내가 그룹바이한 컬럼의 값으로 추출 e.g. pclass면 1,2,3 으로 추출)
----
- 2개 이상으로 그룹 묶는 경우
- 데이터프레임.groupby(['컬럼1','컬럼2'])

----
- 내가 원하는 컬럼을 추출해서 -> grouby로 요약 통계치를 만들자!
- 전체 데이터프레임에서 내가 원하는 데이터만 추출하는 동시에 같이 groupby를 사용하는 것

----
- 여러 개의 통계치를 볼 때 agg함수를 같이 사용해서 진행한다.

- 데이터프레임.groupby('컬럼1').agg(내가원하는 파생변수 명 = ('집계하고 싶은 컬럼1',' 집계하고 싶은 통계치1'), ('집계하고 싶은 컬럼2',' 집계하고 싶은 통계치2'),
-----
- 우리가 원하는 함수를 만들어서 agg함수에 대입시킬 수 있다.

```python
tt.groupby('pclass').agg(fare_mean =('fare','mean'),
                        fare_sum = ('fare','sum'),
                        fare_max = ('fare','max'),
                        fare_median = ('fare','median'))
```

```python
tt.groupby(['sex','pclass']).agg(fare_mean =('fare','mean'),
                        fare_sum = ('fare','sum'),
                        fare_max = ('fare','max'),
                        fare_median = ('fare','median'),
                        fare_count = ('fare','count')) # count 카운팅의 개념
```

```python
tt.groupby('pclass').agg(age_mean = ('age', 'mean'), age_max = ('age', 'max'))
```

```python
def mean_median(x):
    return x.mean()- x.median()
```

```python
## 특정 컬럼을 바로 추출해서 위의 함수를 적용시키는 것

tt.groupby('pclass')['fare'].agg(mean_median)  # fare.mean - fare.median
```
